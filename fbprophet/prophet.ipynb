```python
# Python
import pandas as pd
import numpy as np
from fbprophet import Prophet
```

```python
# 读取文件
df = pd.read_csv('./simple.csv',encoding='utf-8')
# df.rename(columns={'Sales':'y',"Month":'ds'},inplace=True)
# 读取指定的行，我们只需要日期和电力总量
# df.head()

#newdf = pd.DataFrame()
#newdf['df'] = df['日期']
#newdf['y'] = df['用电总和']
df.head(5)
```

```python
m = Prophet()
m.fit(df)
```

```python
# 数据预测，periods表示预测点数，freq表示预测频率
future = m.make_future_dataframe(periods=3,freq='YS') # 预测未来
forecast = m.predict(future)
m.plot(forecast) # 绘制图形
print(forecast)
```
